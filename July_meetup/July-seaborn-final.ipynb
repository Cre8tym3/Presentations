## Pairplot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
g = sns.pairplot(data=iris, kind="scatter")
```

```python
g = sns.pairplot(data=iris, kind="scatter", hue="species")
```

```python
g = sns.pairplot(data=iris, kind="scatter", hue="species", markers=['o', 's', 'D'])
```

```python
g = sns.pairplot(data=iris, kind="scatter", 
                x_vars=["sepal_length", "sepal_width"],
                y_vars=["petal_length", "petal_width"])
```

```python
g = sns.pairplot(data=iris, kind="scatter", 
                diag_kind="kde")
```

```python
g = sns.pairplot(data=iris, kind="reg")
```

```python
g = sns.pairplot(data=iris, kind="reg", hue="species", diag_kind="kde")
```

```python
g = sns.pairplot(data=iris, kind="reg", hue="species", diag_kind="kde", diag_kws={'shade':True})
```

Just like with JointPlot/JointGrid, we have Pair**Grid** which gives us more control.

```python
g = sns.PairGrid(data=iris, hue="species")
g.map_upper(sns.regplot)
g.map_lower(sns.residplot)
g.map_diag(plt.hist)
g.add_legend()
```

## Heatmaps

```python
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
```

```python
ax = sns.heatmap(uniform_data, cmap='Blues')
```

```python
normal_data = np.random.randn(10, 12)
ax = sns.heatmap(normal_data, cmap='Blues')
```

```python
ax = sns.heatmap(normal_data, cmap='coolwarm', robust=True)
```
